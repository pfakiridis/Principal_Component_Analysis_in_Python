```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from pca import pca
data = pd.read_csv("protein.csv")
data.head(5)
```

```python
# Load pca
from pca import pca

# Initialize to reduce the data up to the nubmer of componentes that explains 95% of the variance.
model = pca(n_components=0.95)

# Reduce the data towards 3 PCs
model = pca(n_components=3)

# Fit transform excluding the categorical first column with the country names
results = model.fit_transform(data.iloc[:, 1:])
```

```python
#The scatterplot below give a first impression on how the different countries interact to each other

fig, ax = model.scatter()
```

```python
#The biplot below helps understand how the different variable interact to each other.
#For example it is clear that when the Nuts and Cereals for up the Red Meat and white meat goes down.

fig, ax = model.biplot(n_feat=9)
```

```python
#The below plot helps us to understand the explained variance.
#It is visible that the first PC covers almost 70% of the total variance.
fig, ax = model.plot()
```

```python
#A three dimensional space with the variable

fig, ax = model.biplot3d(n_feat=9)
```
